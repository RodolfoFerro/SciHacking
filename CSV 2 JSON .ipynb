```python
import numpy as np
import pandas as pd

file = pd.read_csv("data1.csv", ",", encoding="latin_1")
df = pd.DataFrame(file)
df.head()
```

```python
campos = {}
campos["OTROS"] = 0
for campo in df["CAMPO"]:
    if campo == "-":
        campos["OTROS"] += 1
    else:
        try:
            campos[campo] += 1
        except:
            campos[campo] = 1
campos["OTROS"] += campos['5'] + campos['7']
del campos['5']
del campos['7']
campos
```

```python
total_campos = 0
for key in campos.keys():
    total_campos += campos[key]
total_campos
```

```python
disciplinas = campos
for key in campos.keys():
    disciplinas[key] = { df["DISCIPLINA"][i]:1 for i in range(len(df["DISCIPLINA"])) if df["CAMPO"][i] == key }
    
for i in range(len(df["DISCIPLINA"])):
    for d_key in disciplinas.keys():
        for key in disciplinas[d_key]:
            if df["DISCIPLINA"][i] == key:
                disciplinas[d_key][key] += 1

disciplinas
```

```python
f = open("test.json", "w")
f.write('{\n\t"name": "CIENCIAS",\n\t"children": [')
for d_key in disciplinas.keys():
    f.write('\n\t\t{\n\t\t\t"name": "')
    f.write(str(d_key.encode('utf-8')))
    f.write('",\n\t\t\t"children": [')
    for key in disciplinas[d_key]:
        f.write('\n\t\t\t{\n\t\t\t\t"name": "')
        f.write(str(key.encode('utf-8')))
        f.write('",\n\t\t\t\t"size": ')
        f.write(str(disciplinas[d_key][key]))
        f.write('\n\t\t\t},')
    f.write('\t\t\t]\n\t\t},')
f.write('\t]\n}')
f.close()
```
